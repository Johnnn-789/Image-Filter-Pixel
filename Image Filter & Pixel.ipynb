```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
img2 = mpimg.imread('photo1.jpg')
bright_img2 = img2 + 50
bright_img2 = np.clip(bright_img2, 0, 255)

img2_norm = img2 / 255.0
gamma = 2.2
gamma_corrected = np.power(img2_norm, gamma)
gamma_corrected = (gamma_corrected * 255).astype(np.uint8)

vintage = np.sqrt(img2_norm)
vintage = (vintage * 255).astype(np.uint8)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(bright_img2.astype(np.uint8))
plt.title("Brightened")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(gamma_corrected)
plt.title("Gamma Corrected")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(vintage)
plt.title("Vintage Effect")
plt.axis('off')
plt.show()
background = np.zeros_like(img2)  
mask = (img2[:, :, 1] > 150) & (img2[:, :, 1] > img2[:, :, 0]) & (img2[:, :, 1] > img2[:, :, 2])
result = img2.copy()
result[mask] = background[mask]
plt.imshow(result)
plt.title("Masked Background Replacement")
plt.axis('off')
plt.show()
mask_labels = np.random.randint(0, 4, size=(img2.shape[0], img2.shape[1]))  # fake labels
colors = {
    0: (0, 0, 0),
    1: (255, 0, 0),
    2: (0, 255, 0),
    3: (0, 0, 255)
}
colored = np.zeros_like(img2)
for label, color in colors.items():
    colored[mask_labels == label] = color
plt.imshow(colored)
plt.title("Colored Labels")
plt.axis('off')
plt.show()
```
